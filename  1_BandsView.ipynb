```python
src = rio.open('20210523_RGB.tif')
from rasterio.plot import show
show(src)
```

```python
fig, (axn, axr, axg, axb, axa ) = plt.subplots(1,5, figsize=(20,10))
show((src, 5), ax=axn, cmap='Greys_r', title='NIR')
show((src, 4), ax=axr, cmap='Greys_r', title='red channel')
show((src, 3), ax=axg, cmap='Greys_r', title='green channel')
show((src, 2), ax=axb, cmap='Greys_r', title='blue channel')
show((src, 1), ax=axa, cmap='Greys_r', title='aerosol')
plt.show()
```

```python
fig, (axn, axr, ) = plt.subplots(1,2, figsize=(15,10))
show((src, 5), ax=axn, cmap='Greys_r', title='NIR')
show((src, 4), ax=axr, cmap='Reds', title='red channel')
plt.show()
fig, (axg, axb, axa ) = plt.subplots(1,3, figsize=(19,10))
show((src, 3), ax=axg, cmap='Greens', title='green channel')
show((src, 2), ax=axb, cmap='Blues', title='blue channel')
show((src, 1), ax=axa, cmap='Greys_r', title='Coastal aerosol')
plt.show()
```

```python
fig, (axa, axb, axg ) = plt.subplots(1,3, figsize=(19,10))
show((src, 1), cmap='Greys_r', ax=axa)
show((src, 1), contour=True, ax=axa ,title='Coastal aerosol')
show((src, 2), cmap='Greys_r', ax=axb , title='Blue channel')
show((src, 2), contour=True, ax=axb)
show((src, 3), cmap='Greys_r', ax=axg , title='Green channel')
show((src, 3), contour=True, ax=axg)
plt.show()
```

```python
fig, (axn, axr ) = plt.subplots(1,2, figsize=(10,10))
show((src, 4), cmap='Greys_r', ax=axr , title='red channel')
show((src, 4), contour=True, ax=axr)
show((src, 5), cmap='Greys_r', ax=axn)
show((src, 5), contour=True, ax=axn ,title='NIR channel')
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(12, 12))
show((src, 1), cmap='Greys_r', interpolation='none', ax=ax)
show((src, 1), contour=True, ax=ax)
plt.show()
```

```python
from rasterio.plot import show_hist
show_hist(src, bins=40, lw=0.0, stacked=False, alpha=0.3,
          histtype='stepfilled', title="Histogram")
```
